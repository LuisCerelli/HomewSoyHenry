```python
archivo = open('Emisiones_CO2.csv', "r", encoding= "windows-1252")
```

```python
dicc_emisiones = { 'cod_pais': [],
                  'nom_pais': [],
                  'region': [],
                  'anio': [],
                  'co2': [],
                  'co2_percapita': []
                  }
```

```python
next(archivo)
for linea in archivo:
    campos = linea.strip().split("|")
    dicc_emisiones["cod_pais"].append(campos[0])
    dicc_emisiones["nom_pais"].append(campos[1])
    dicc_emisiones["region"].append(campos[2])
    dicc_emisiones["anio"].append(campos[3])
    dicc_emisiones["co2"].append(campos[4])
    dicc_emisiones["co2_percapita"].append(campos[5])
```

d) ¿Hay valores faltantes? <br>

```python
def contar_nulos(variable):
    nulo = 0
    for elemento in variable:
        if elemento == "":
            nulo += 1
    return nulo
```

```python
for llave in dicc_emisiones.keys():
    nulos = contar_nulos(dicc_emisiones[llave])
    print(f"La cantidad de nulos para la llave {llave}es: {nulos}")
```

e) ¿Cuál es el total de emisiones de CO2 para 'América Latina y Caribe' en el año 2010?

```python
nuevos_valores_co2 = []
for valor in dicc_emisiones["co2"]:
    if valor: #que es lo mismo que colocar if valor == True, o sea, que el valor existe, no está vacio:
        nuevo_valor = float(valor.replace(".", "").replace(",", "."))
    else:
        nuevo_valor = None
    nuevos_valores_co2.append(nuevo_valor)
dicc_emisiones["co2"] = nuevos_valores_co2
```
